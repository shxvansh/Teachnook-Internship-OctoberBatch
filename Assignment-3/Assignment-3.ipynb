```python
#simple calculator

num1 = float(input("Enter your first number:"))
num2 = float(input("Enter your second number:"))
op = input("Enter your operator:")

def add():
    print("sum of you numbers is:",num1+num2)
def subtract():
    print("diff of your numbers is:",num1-num2)
def mult():
    print("multiplication of %s and %s is %s"%(num1,num2,(num1*num2)))
def div():
    print("division of %s and %s is %s"%(num1,num2,num1/num2))
def power():
    print("%s to the power of %s is %s"%(num1,num2,num1**num2))
    
if (op == "+"):
    add()
elif (op == "-"):
    subtract()
elif (op == "*"):
    mult()
elif (op == "**"):
    power()
elif (op == ("/")):
    div()
else:
    print("choose the correct operand next time.")

```
