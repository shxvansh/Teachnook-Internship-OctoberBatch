```python
#set
#unordered collection of datatypes
#symbol used to enclose {}
#no duplicates allowed

```

```python
set = {1,2,3,4,5}
print(set)
```

```python
#adding element to a set
set.add(6)
print(set)
```

```python
#deleting element from a set
set.remove(6)
print(set)
```

Sets are mutable ie we can add or remove items from the set.
